- **Menghandle Outlier**

```python
# Untuk mengecek apakah penggabungan berhasil:
print("Shape air_quality_df:", air_quality_df.shape)
print("\nSample data:")
print(air_quality_df.head())

# Kemudian lanjutkan dengan handling outlier
outlier_feature = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

# Tambahkan print untuk debugging
print("\nBefore outlier handling:")
print(air_quality_df[outlier_feature].describe())

Q1 = air_quality_df[outlier_feature].quantile(0.25)
Q3 = air_quality_df[outlier_feature].quantile(0.75)
IQR = Q3 - Q1
maximum = Q3 + (1.5 * IQR)
minimum = Q1 - (1.5 * IQR)

for col in outlier_feature:
    kondisi_lower_than = air_quality_df[col] < minimum[col]
    kondisi_more_than = air_quality_df[col] > maximum[col]

    air_quality_df[col] = air_quality_df[col].where(~kondisi_more_than, maximum[col])
    air_quality_df[col] = air_quality_df[col].where(~kondisi_lower_than, minimum[col])

# Tambahkan print untuk memastikan perubahan
print("\nAfter outlier handling:")
print(air_quality_df[outlier_feature].describe())
```

```python
# Membuat kolom bulan dan tahun dari datetime
air_quality_df['month_year'] = air_quality_df['datetime'].dt.strftime('%b %Y')
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(15, 15))

# Dictionary untuk konfigurasi plot
plots_config = [
    {'param': 'SO2', 'title': 'Jumlah SO2 per Periode Waktu'},
    {'param': 'NO2', 'title': 'Jumlah NO2 per Periode Waktu'},
    {'param': 'CO', 'title': 'Jumlah CO per Periode Waktu'},
    {'param': 'O3', 'title': 'Jumlah O3 per Periode Waktu'}
]

# Membuat plot untuk setiap parameter
for idx, config in enumerate(plots_config):
    sns.barplot(
        x='month_year',
        y=config['param'],
        data=air_quality_df,
        palette='rocket',
        ax=axes[idx]
    )

    # Konfigurasi setiap subplot
    axes[idx].set_title(config['title'], pad=15)
    axes[idx].set_xlabel('Periode (Bulan-Tahun)')
    axes[idx].set_ylabel(f"{config['param']} (g/m)")

    # Memperbaiki alignment label
    axes[idx].set_xticklabels(
        axes[idx].get_xticklabels(),
        rotation=45,
        horizontalalignment='right'
    )

    # Menambahkan grid untuk memudahkan pembacaan
    axes[idx].grid(True, axis='y', linestyle='--', alpha=0.3)

    # Mengatur margin subplot
    axes[idx].margins(x=0.01)

# Mengatur layout keseluruhan
plt.tight_layout(h_pad=3.0)  # Menambah jarak antar subplot

# Menampilkan plot
plt.show()
```

- Jumlah NO2 tertinggi tertuju pada bulan Desember tahun 2016 dan yang terendah adalah bulan September tahun 2015
- Jumlah CO tertinggi tertuju pada bulan Desember tahun 2015 dan yang terendah adalah bulan Mei tahun 2016
- Jumlah O3 tertinggi tertuju pada bulan Juni tahun 2016 dan yang terendah adalah bulan November tahun 2015

### Pertanyaan 2: Bagaimana perbedaan PM2.5 dan PM10 Tahun 2013-2017 di seluruh distrik

```python
# Calculate yearly averages for both PM2.5 and PM10
air_quality_df['year'] = air_quality_df['datetime'].dt.year
yearly_avg = air_quality_df.groupby('year').agg({
    'PM2.5': 'mean',
    'PM10': 'mean'
}).reset_index()

# Create the visualization
plt.figure(figsize=(12, 8))

# Plot both lines
plt.plot(yearly_avg['year'], yearly_avg['PM2.5'], marker='o', linewidth=2,
         label='PM2.5', color='#2ecc71', markersize=10, zorder=2)
plt.plot(yearly_avg['year'], yearly_avg['PM10'], marker='s', linewidth=2,
         label='PM10', color='#e74c3c', markersize=10, zorder=2)

# Add labels and title
plt.title('Perbandingan Rata-rata Tahunan PM2.5 dan PM10 (2013-2017)',
          fontsize=14, pad=20)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Konsentrasi (g/m)', fontsize=12)

# Customize the grid and put it behind the data
plt.grid(True, linestyle='--', alpha=0.7, zorder=1)

# Customize x-axis
plt.xticks(yearly_avg['year'], rotation=0)

# Move legend to top right corner of the plot area
plt.legend(fontsize=10, loc='upper right', bbox_to_anchor=(1.0, 1.0))

# Add value labels with offset positions
for i, row in yearly_avg.iterrows():
    # PM2.5 labels below points
    plt.text(row['year'], row['PM2.5'] - 2.5, f'{row["PM2.5"]:.1f}',
             ha='center', va='top', color='#2ecc71', fontweight='bold')
    # PM10 labels above points
    plt.text(row['year'], row['PM10'] + 2.5, f'{row["PM10"]:.1f}',
             ha='center', va='bottom', color='#e74c3c', fontweight='bold')

# Calculate proper y-axis range to show all data
y_min = min(yearly_avg['PM2.5'].min(), yearly_avg['PM10'].min()) - 8
y_max = max(yearly_avg['PM2.5'].max(), yearly_avg['PM10'].max()) + 8

# Set y-axis limits to show all data properly
plt.ylim(y_min, y_max)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

## Pertanyaan 3: Bagaimana rata-rata kualitas udara tahun 2013 - 2017

```python
def categorize_aqi(row):
    pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
    valid_values = []

    for pollutant in pollutants:
        if pollutant in row and pd.notna(row[pollutant]):
            valid_values.append(row[pollutant])

    if not valid_values:
        return None

    return sum(valid_values) / len(valid_values)

# Kategorikan AQI
air_quality_df['aqi'] = air_quality_df.apply(categorize_aqi, axis=1)

def get_category(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

# Kategorikan
air_quality_df['category'] = air_quality_df['aqi'].apply(get_category)

# Hitung persentase untuk setiap kategori
category_counts = air_quality_df['category'].value_counts()
category_percentages = (category_counts / len(air_quality_df) * 100).round(1)

# Warna untuk setiap kategori
colors = {
    'Good': '#00e400',
    'Moderate': '#ffff00',
    'Unhealthy for Sensitive Groups': '#ff7e00',
    'Unhealthy': '#ff0000',
    'Very Unhealthy': '#8f3f97',
    'Hazardous': '#7e0023'
}

# Buat pie chart
plt.figure(figsize=(10, 10))
plt.pie(category_percentages,
        labels=[f'{cat}\n{pct}%' for cat, pct in category_percentages.items()],
        colors=[colors[cat] for cat in category_percentages.index],
        startangle=90)

plt.title('Air Quality Distribution (2013-2017)', pad=20)
plt.axis('equal')
plt.show()
```

## Pertanyaan 4: Bagaimana perbandingan rata-rata kualitas terburuk dan tersehat pada distrik di beijing?

```python
# Hitung AQI dan kategorikan
air_quality_df['aqi'] = air_quality_df.apply(categorize_aqi, axis=1)
air_quality_df['air_quality'] = air_quality_df['aqi'].apply(get_category)

# Hapus data null
air_quality_df = air_quality_df.dropna(subset=['air_quality'])

# Definisikan warna
quality_colors = {
    'Good': '#00e400',
    'Moderate': '#ffff00',
    'Unhealthy for Sensitive Groups': '#ff7e00',
    'Unhealthy': '#ff0000',
    'Very Unhealthy': '#8f3f97',
    'Hazardous': '#7e0023'
}

# Buat visualisasi
plt.figure(figsize=(15, 8))

district_quality = pd.crosstab(air_quality_df['station'],
                              air_quality_df['air_quality'],
                              normalize='index') * 100

district_quality.plot(kind='bar',
                     stacked=True,
                     color=[quality_colors[col] for col in district_quality.columns])

plt.title('Air Quality Distribution by District (2013-2017)', pad=20)
plt.xlabel('District')
plt.ylabel('Percentage')
plt.legend(title='Air Quality', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analisis Lanjutan

Analisis clustering menggunakan beberapa parameter yang relevan dari data kualitas udara tersebut. Untuk kasus ini, kita bisa menggunakan parameter-parameter seperti PM2.5, PM10, SO2, NO2, CO, dan O3 yang merupakan indikator utama kualitas udara.

```python
def get_air_quality_stats(air_quality_df):
    # Pilih parameter yang akan digunakan
    parameters = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

    # Handling missing values
    for param in parameters:
        air_quality_df[param].fillna(air_quality_df[param].mean(), inplace=True)

    # Manual Grouping untuk PM2.5
    def categorize_pm25(value):
        if value <= 12:
            return 'Good'
        elif value <= 35.4:
            return 'Moderate'
        elif value <= 55.4:
            return 'Unhealthy for sensitive groups'
        elif value <= 150.4:
            return 'Unhealthy'
        elif value <= 250.4:
            return 'Very unhealthy'
        else:
            return 'Hazardous'

    air_quality_df['PM2.5_Category'] = air_quality_df['PM2.5'].apply(categorize_pm25)

    # Binning untuk parameter lain
    for param in parameters[1:]:  # Skip PM2.5 karena sudah dikategorikan
        air_quality_df[f'{param}_Binned'] = pd.qcut(air_quality_df[param], q=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'], duplicates='drop')

    # Statistik per kategori PM2.5
    stats_by_category = air_quality_df.groupby('PM2.5_Category')[parameters].agg(['mean', 'std', 'count'])

    # Statistik per distrik
    district_summary = air_quality_df.groupby('station')[parameters].agg(['mean', 'max', 'min'])

    return stats_by_category, district_summary

# Jalankan analisis
category_stats, district_stats = get_air_quality_stats(air_quality_df)

# Tampilkan hasil
print("\nStatistik per Kategori PM2.5:")
print(category_stats)
print("\nStatistik per Distrik:")
print(district_stats)
```

-Lingkaran hitam di atas garis horizontal yang merupakan batas atas outlier adalah banayaknya sample di luar outlier/ lebih dari outlier

**Menilai batas2 outlier di kolom PM 2.5**

```python
# Menghapus nilai NaN di kolom PM2.5
data_cleaned = combined_df["PM2.5"].dropna()

# Menghitung Q1, Q3, dan IQR
Q1 = data_cleaned.quantile(0.25)
Q3 = data_cleaned.quantile(0.75)
IQR = Q3 - Q1

# Menghitung batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengidentifikasi outlier
outliers = data_cleaned[(data_cleaned < lower_bound) | (data_cleaned > upper_bound)]

# Menampilkan hasil
print(f"Jumlah outlier pada kolom PM2.5: {len(outliers)}")
print(f"Batas bawah: {lower_bound}, Batas atas: {upper_bound}")
```

```python
import pandas as pd

Distrik_Aotizhongxin = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv')
Distrik_Changping = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Changping_20130301-20170228.csv')
Distrik_Dingling = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Dingling_20130301-20170228.csv')
Distrik_Dongsi = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Dongsi_20130301-20170228.csv')
Distrik_Guanyuan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Guanyuan_20130301-20170228.csv')
Distrik_Gucheng = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Gucheng_20130301-20170228.csv')
Distrik_Huairou = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Huairou_20130301-20170228.csv')
Distrik_Nongzhanguan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Nongzhanguan_20130301-20170228.csv')
Distrik_Shunyi = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Shunyi_20130301-20170228.csv')
Distrik_Tiantan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Tiantan_20130301-20170228.csv')
Distrik_Wanliu = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Wanliu_20130301-20170228.csv')
Distrik_Wanshouxigong = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Wanshouxigong_20130301-20170228.csv')

# Kemudian gabungkan
df_merged = [Distrik_Aotizhongxin, Distrik_Changping, Distrik_Dingling,
           Distrik_Dongsi, Distrik_Guanyuan, Distrik_Gucheng,
           Distrik_Huairou, Distrik_Nongzhanguan, Distrik_Shunyi,
           Distrik_Tiantan, Distrik_Wanliu, Distrik_Wanshouxigong]

combined_df = pd.concat(df_merged, ignore_index=True)

# Untuk mengecek apakah penggabungan berhasil:
print("Shape combined_df:", combined_df.shape)
print("\nSample data:")
print(combined_df.head())

# Kemudian lanjutkan dengan handling outlier
outlier_feature = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

# Tambahkan print untuk debugging
print("\nBefore outlier handling:")
print(combined_df[outlier_feature].describe())

Q1 = combined_df[outlier_feature].quantile(0.25)
Q3 = combined_df[outlier_feature].quantile(0.75)
IQR = Q3 - Q1
maximum = Q3 + (1.5 * IQR)
minimum = Q1 - (1.5 * IQR)

for col in outlier_feature:
    kondisi_lower_than = combined_df[col] < minimum[col]
    kondisi_more_than = combined_df[col] > maximum[col]

    combined_df[col] = combined_df[col].where(~kondisi_more_than, maximum[col])
    combined_df[col] = combined_df[col].where(~kondisi_lower_than, minimum[col])

# Tambahkan print untuk memastikan perubahan
print("\nAfter outlier handling:")
print(combined_df[outlier_feature].describe())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

Distrik_Aotizhongxin = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv')
Distrik_Changping = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Changping_20130301-20170228.csv')
Distrik_Dingling = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Dingling_20130301-20170228.csv')
Distrik_Dongsi = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Dongsi_20130301-20170228.csv')
Distrik_Guanyuan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Guanyuan_20130301-20170228.csv')
Distrik_Gucheng = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Gucheng_20130301-20170228.csv')
Distrik_Huairou = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Huairou_20130301-20170228.csv')
Distrik_Nongzhanguan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Nongzhanguan_20130301-20170228.csv')
Distrik_Shunyi = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Shunyi_20130301-20170228.csv')
Distrik_Tiantan = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Tiantan_20130301-20170228.csv')
Distrik_Wanliu = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Wanliu_20130301-20170228.csv')
Distrik_Wanshouxigong = pd.read_csv('https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Wanshouxigong_20130301-20170228.csv')

df_merged = [Distrik_Aotizhongxin, Distrik_Changping, Distrik_Dingling, Distrik_Dongsi, Distrik_Guanyuan, Distrik_Gucheng, Distrik_Huairou, Distrik_Nongzhanguan, Distrik_Shunyi, Distrik_Tiantan, Distrik_Wanliu, Distrik_Wanshouxigong]
air_quality_merge = pd.concat(df_merged)

def get_air_quality_stats(df_merged):
    # Pilih parameter yang akan digunakan
    parameters = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

    # Handling missing values
    for param in parameters:
        df_merged[param].fillna(df_merged[param].mean(), inplace=True)

    # Manual Grouping untuk PM2.5
    def categorize_pm25(value):
        if value <= 12:
            return 'Good'
        elif value <= 35.4:
            return 'Moderate'
        elif value <= 55.4:
            return 'Unhealthy for sensitive groups'
        elif value <= 150.4:
            return 'Unhealthy'
        elif value <= 250.4:
            return 'Very unhealthy'
        else:
            return 'Hazardous'

    df_merged['PM2.5_Category'] = df_merged['PM2.5'].apply(categorize_pm25)

    # Binning untuk parameter lain
    for param in parameters[1:]:  # Skip PM2.5 karena sudah dikategorikan
        df_merged[f'{param}_Binned'] = pd.qcut(df_merged[param], q=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

    # Statistik per kategori PM2.5
    stats_by_category = df_merged.groupby('PM2.5_Category')[parameters].agg(['mean', 'std', 'count'])

    # Statistik per distrik
    district_summary = df_merged.groupby('station')[parameters].agg(['mean', 'max', 'min'])

    return stats_by_category, district_summary

# Jalankan analisis
category_stats, district_stats = get_air_quality_stats(air_quality_merge)

# Tampilkan hasil
print("\nStatistik per Kategori PM2.5:")
print(category_stats)
print("\nStatistik per Distrik:")
print(district_stats)
```

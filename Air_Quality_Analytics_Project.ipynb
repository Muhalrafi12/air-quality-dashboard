- **Menghandle Outlier**

```python
# Kemudian lanjutkan dengan handling outlier
all_pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
Q1 = air_quality_df[all_pollutants].quantile(0.25)
Q3 = air_quality_df[all_pollutants].quantile(0.75)
IQR = Q3 - Q1
maximum = Q3 + (1.5 * IQR)
minimum = Q1 - (1.5 * IQR)

for col in all_pollutants:
    kondisi_lower_than = air_quality_df[col] < minimum[col]
    kondisi_more_than = air_quality_df[col] > maximum[col]

    air_quality_df[col] = air_quality_df[col].where(~kondisi_more_than, maximum[col])
    air_quality_df[col] = air_quality_df[col].where(~kondisi_lower_than, minimum[col])
```

- **Memformat Tipe Datetime**
    - Mengubah tipe data 'year', 'month', 'day', 'hour' menjadi datetime

```python
if 'year' in air_quality_df.columns and 'month' in air_quality_df.columns and 'day' in air_quality_df.columns and 'hour' in air_quality_df.columns:
    air_quality_df['datetime'] = pd.to_datetime(air_quality_df[['year', 'month', 'day', 'hour']])

    air_quality_df['year'] = air_quality_df['datetime'].dt.year
    air_quality_df['month'] = air_quality_df['datetime'].dt.month
    air_quality_df['day'] = air_quality_df['datetime'].dt.day
    air_quality_df['hour'] = air_quality_df['datetime'].dt.hour
    air_quality_df['month_year'] = air_quality_df['datetime'].dt.strftime('%Y-%m')
else:
    if 'datetime' not in air_quality_df.columns:
        pass
    if 'year' not in air_quality_df.columns and 'datetime' in air_quality_df.columns:
        air_quality_df['year'] = air_quality_df['datetime'].dt.year
    if 'month' not in air_quality_df.columns and 'datetime' in air_quality_df.columns:
        air_quality_df['month'] = air_quality_df['datetime'].dt.month
    if 'day' not in air_quality_df.columns and 'datetime' in air_quality_df.columns:
        air_quality_df['day'] = air_quality_df['datetime'].dt.day
    if 'hour' not in air_quality_df.columns and 'datetime' in air_quality_df.columns:
        air_quality_df['hour'] = air_quality_df['datetime'].dt.hour
    if 'month_year' not in air_quality_df.columns and 'datetime' in air_quality_df.columns:
        air_quality_df['month_year'] = air_quality_df['datetime'].dt.strftime('%Y-%m')
```

```python
def calculate_aqi(row):
    valid_values = []
    for pollutant in all_pollutants:
        if pollutant in row and pd.notna(row[pollutant]):
            valid_values.append(row[pollutant])
    if not valid_values:
        return None
    return sum(valid_values) / len(valid_values)

def categorize_air_quality(aqi):
    if aqi <= 50: return 'Good'
    elif aqi <= 100: return 'Moderate'
    elif aqi <= 150: return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200: return 'Unhealthy'
    elif aqi <= 300: return 'Very Unhealthy'
    else: return 'Hazardous'

air_quality_df['aqi'] = air_quality_df.apply(calculate_aqi, axis=1)
air_quality_df['category'] = air_quality_df['aqi'].apply(categorize_air_quality)

# Distribusi kategori
category_counts = air_quality_df['category'].value_counts()
category_pct = (category_counts / len(air_quality_df) * 100).round(2)

pd.DataFrame({
    'Count': category_counts,
    'Percentage': category_pct
})

# Distribusi per tahun
pd.crosstab(air_quality_df['year'], air_quality_df['category'], normalize='index') * 100

# Statistik AQI per kategori
air_quality_df.groupby('category')['aqi'].describe()
```

Tekanan udara (PRES) relatif stabil (std: 10.5 hPa) dengan rata-rata 1010.7 hPa, namun nilai maksimum 1042.8 hPa mengindikasikan episode inversi termal yang menahan polusi. Titik embun (DEWP) sangat bervariasi (-43.4C hingga 29.1C) dengan std tinggi (13.8C), mencerminkan fluktuasi kelembaban ekstrem. Data ini mengkonfirmasi kondisi meteorologi dominan tidak mendukung dispersi alami polusi - kombinasi angin lemah, hujan jarang, dan variabilitas suhu tinggi menciptakan lingkungan rentan akumulasi polutan.

```python
# 3. Pengaruh suhu terhadap polutan (binning suhu)
air_quality_df['temp_category'] = pd.cut(
    air_quality_df['TEMP'],
    bins=[-50, 0, 10, 20, 30, 50],
    labels=['Very Cold (<0C)', 'Cold (0-10C)', 'Moderate (10-20C)', 'Warm (20-30C)', 'Hot (>30C)']
)

temp_impact = air_quality_df.groupby('temp_category')[all_pollutants].mean()
temp_impact.round(2)
```

```python
# 4. Pengaruh kecepatan angin terhadap polutan
air_quality_df['wind_category'] = pd.cut(
    air_quality_df['WSPM'],
    bins=[0, 1, 3, 5, 20],
    labels=['Calm (0-1 m/s)', 'Light (1-3 m/s)', 'Moderate (3-5 m/s)', 'Strong (>5 m/s)']
)

wind_impact = air_quality_df.groupby('wind_category')[all_pollutants].mean()
wind_impact.round(2)
```

**Insight**
- Suhu dingin (<10C) adalah kondisi terburuk dengan PM2.5 (82-86 g/m) dan CO (1320-1552 g/m) tertinggi akibat inversi termal dan emisi pemanas, sementara suhu panas (>30C) memicu lonjakan O3 drastis (146 g/m) dari reaksi fotokimia namun menurunkan polutan primer hingga 50%. Pola ini mengkonfirmasi musim dingin memerlukan pembatasan emisi ketat, sedangkan musim panas butuh mitigasi khusus ozon meski kualitas udara umum lebih baik.

```python
# 5. Pengaruh hujan terhadap polutan (hari hujan vs tidak hujan)
air_quality_df['is_rainy'] = air_quality_df['RAIN'].apply(lambda x: 'Rainy Day' if x > 0 else 'No Rain')

rain_impact = air_quality_df.groupby('is_rainy')[all_pollutants].mean()
rain_impact.round(2)
```

**Insight**
- Hujan tiba dapat menyebabkan impact yang besar untuk polutan PM10, SO2, NO2, namun untuk polutan lainnya juga berimpact namun tidak terlalu signifikan

```python
# 6. Kombinasi kondisi meteorologi ekstrem
air_quality_df['meteo_condition'] = 'Normal'
air_quality_df.loc[(air_quality_df['WSPM'] < 1) & (air_quality_df['TEMP'] < 5), 'meteo_condition'] = 'High Risk (Cold+Calm)'
air_quality_df.loc[(air_quality_df['WSPM'] > 5), 'meteo_condition'] = 'Low Risk (Windy)'
air_quality_df.loc[(air_quality_df['TEMP'] > 25) & (air_quality_df['WSPM'] < 2), 'meteo_condition'] = 'O3 Risk (Hot+Calm)'

meteo_condition_impact = air_quality_df.groupby('meteo_condition')[all_pollutants + ['aqi']].mean()
meteo_condition_impact.round(2)
```

**Insight**
- Kondisi "High Risk (Cold+Calm)" adalah skenario paling berbahaya dengan polutan primer mencapai level kritis dan AQI kategori Very Unhealthy, mengkonfirmasi inversi termal dan stagnasi udara menyebabkan akumulasi polusi ekstrem. Sebaliknya, "Low Risk (Windy)" menunjukkan penurunan dramatis semua polutan dengan AQI turun ke kategori Moderate, membuktikan angin kencang sebagai mekanisme dispersi paling efektif.
"O3 Risk (Hot+Calm)" menghadirkan bahaya berbeda - meski polutan primer berkurang signifikan, ozon melonjak ke level Unhealthy akibat reaksi fotokimia intensif. Kondisi "Normal" berada di tengah dengan AQI Unhealthy. Strategi mitigasi harus dibedakan: pembatasan emisi darurat saat dingin+tenang untuk polutan primer, dan peringatan kesehatan khusus saat panas+tenang untuk paparan ozon.Claude adalah AI dan bisa keliru.

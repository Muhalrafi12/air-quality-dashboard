- **Menghandle Outlier**

```python
# Untuk mengecek apakah penggabungan berhasil:
print("Shape air_quality_df:", air_quality_df.shape)
print("\nSample data:")
print(air_quality_df.head())

# Kemudian lanjutkan dengan handling outlier
outlier_feature = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

# Tambahkan print untuk debugging
print("\nBefore outlier handling:")
print(air_quality_df[outlier_feature].describe())

Q1 = air_quality_df[outlier_feature].quantile(0.25)
Q3 = air_quality_df[outlier_feature].quantile(0.75)
IQR = Q3 - Q1
maximum = Q3 + (1.5 * IQR)
minimum = Q1 - (1.5 * IQR)

for col in outlier_feature:
    kondisi_lower_than = air_quality_df[col] < minimum[col]
    kondisi_more_than = air_quality_df[col] > maximum[col]

    air_quality_df[col] = air_quality_df[col].where(~kondisi_more_than, maximum[col])
    air_quality_df[col] = air_quality_df[col].where(~kondisi_lower_than, minimum[col])

# Tambahkan print untuk memastikan perubahan
print("\nAfter outlier handling:")
print(air_quality_df[outlier_feature].describe())
```

### Pertanyaan 1: Kapan jumlah polutan SO2, NO2, CO, dan O3 mencapai titik tertinggi

```python
# Membuat kolom bulan dan tahun dari datetime
air_quality_df['month_year'] = air_quality_df['datetime'].dt.strftime('%b %Y')
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(15, 15))

# Dictionary untuk konfigurasi plot
plots_config = [
    {'param': 'SO2', 'title': 'Jumlah SO2 per Periode Waktu'},
    {'param': 'NO2', 'title': 'Jumlah NO2 per Periode Waktu'},
    {'param': 'CO', 'title': 'Jumlah CO per Periode Waktu'},
    {'param': 'O3', 'title': 'Jumlah O3 per Periode Waktu'}
]

# Membuat plot untuk setiap parameter
for idx, config in enumerate(plots_config):
    sns.barplot(
        x='month_year',
        y=config['param'],
        data=air_quality_df,
        palette='rocket',
        ax=axes[idx]
    )

    # Konfigurasi setiap subplot
    axes[idx].set_title(config['title'], pad=15)
    axes[idx].set_xlabel('Periode (Bulan-Tahun)')
    axes[idx].set_ylabel(f"{config['param']} (g/m)")

    # Memperbaiki alignment label
    axes[idx].set_xticklabels(
        axes[idx].get_xticklabels(),
        rotation=45,
        horizontalalignment='right'
    )

    # Menambahkan grid untuk memudahkan pembacaan
    axes[idx].grid(True, axis='y', linestyle='--', alpha=0.3)

    # Mengatur margin subplot
    axes[idx].margins(x=0.01)

# Mengatur layout keseluruhan
plt.tight_layout(h_pad=3.0)  # Menambah jarak antar subplot

# Menampilkan plot
plt.show()
```

Hasil analisis:
- Jumlah SO2 tertinggi tertuju pada bulan Januari-Februari tahun 2014 dan yang terendah adalah bulan Agustus tahun 2016
- Jumlah NO2 tertinggi tertuju pada bulan Desember tahun 2016 dan yang terendah adalah bulan September tahun 2015
- Jumlah CO tertinggi tertuju pada bulan Desember tahun 2015 dan yang terendah adalah bulan Mei tahun 2016
- Jumlah O3 tertinggi tertuju pada bulan Juni tahun 2016 dan yang terendah adalah bulan November tahun 2015

Conclution

Berdasarkan analisis tren bulanan, terdapat perbedaan pola musiman yang kontras antara O3 dengan polutan lainnya (SO2, NO2, CO).

## Pertanyaan 3:  Bagaimana distribusi kualitas udara pada rentang waktu 2013-2017

```python
def categorize_aqi(row):
    pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
    valid_values = []

    for pollutant in pollutants:
        if pollutant in row and pd.notna(row[pollutant]):
            valid_values.append(row[pollutant])

    if not valid_values:
        return None

    return sum(valid_values) / len(valid_values)

# Kategorikan AQI
air_quality_df['aqi'] = air_quality_df.apply(categorize_aqi, axis=1)

def get_category(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

# Kategorikan
air_quality_df['category'] = air_quality_df['aqi'].apply(get_category)

# Hitung persentase untuk setiap kategori
category_counts = air_quality_df['category'].value_counts()
category_percentages = (category_counts / len(air_quality_df) * 100).round(1)

# Warna untuk setiap kategori
colors = {
    'Good': '#00e400',
    'Moderate': '#ffff00',
    'Unhealthy for Sensitive Groups': '#ff7e00',
    'Unhealthy': '#ff0000',
    'Very Unhealthy': '#8f3f97',
    'Hazardous': '#7e0023'
}

# Buat pie chart
plt.figure(figsize=(10, 10))
plt.pie(category_percentages,
        labels=[f'{cat}\n{pct}%' for cat, pct in category_percentages.items()],
        colors=[colors[cat] for cat in category_percentages.index],
        startangle=90)

plt.title('Air Quality Distribution (2013-2017)', pad=20)
plt.axis('equal')
plt.show()
```

Konklusi:
- Berdasarkan analisis distribusi kategori kualitas udara dari tahun 2013 hingga 2017, ditemukan fakta yang mengkhawatirkan. Kategori kualitas udara yang paling dominan adalah 'Hazardous' (Berbahaya) dengan persentase tertinggi mencapai 26.7%. Sebaliknya, kategori 'Good' (Baik) justru menjadi minoritas terkecil dengan persentase hanya 2.2%.

Jika kategori 'Hazardous', 'Very Unhealthy', dan 'Unhealthy' digabungkan, mereka mendominasi lebih dari 60% total hari dalam rentang waktu tersebut. Hal ini mengindikasikan bahwa kondisi udara buruk di Beijing bukanlah kejadian sesekali (anomali), melainkan sudah menjadi kondisi 'normal' atau kronis yang terjadi hampir setiap hari akibat beban emisi yang jauh melebihi kapasitas lingkungan.

Mengingat kategori 'Good' sangat langka (hanya 2.2%), maka pendekatan 'reaktif' (bertindak saat ada peringatan) tidak lagi efektif. Berikut rekomendasi strategisnya:

1. Standar Operasional Prosedur (SOP) Perkantoran & Sekolah: Institusi tidak boleh menunggu peringatan manual. Disarankan menerapkan kebijakan Work From Home (WFH) atau sekolah daring otomatis jika sensor menunjukkan kategori masuk ke level 'Hazardous' (26.7% kemungkinan terjadi), demi menjaga kesehatan jangka panjang karyawan dan siswa.

2. Investasi Infrastruktur Filtrasi Udara: Bagi pengelola gedung komersial dan fasilitas umum, pemasangan sistem HEPA Filter industri bukan lagi opsi tambahan, melainkan kebutuhan primer. Peluang bisnis penyediaan udara bersih dalam ruangan menjadi sangat krusial mengingat 97.8% waktu udara luar tidak dalam kondisi 'Good'.

3. Kampanye 'Masker sebagai Kebutuhan Pokok': Pemerintah perlu menggeser narasi penggunaan masker N95 dari 'alat darurat' menjadi 'pakaian sehari-hari' saat beraktivitas di luar ruangan, mengingat peluang mendapatkan udara bersih kurang dari 3%.

## Pertanyaan 4: Bagaimana distribusi tingkat kualitas udara di berbagai distrik Beijing pada periode 2013-2017?

```python
# --- 1. Persiapan Data ---
# Hitung AQI dan kategorikan (Pastikan fungsi categorize_aqi dan get_category sudah di-run di cell sebelumnya)
air_quality_df['aqi'] = air_quality_df.apply(categorize_aqi, axis=1)
air_quality_df['air_quality'] = air_quality_df['aqi'].apply(get_category)

# Hapus data null pada kategori
air_quality_df = air_quality_df.dropna(subset=['air_quality'])

# --- 2. Definisi Urutan & Warna ---
# PENTING: Kita tentukan urutan kategori agar di chart 'Good' ada di bawah dan 'Hazardous' di atas (urut)
category_order = [
    'Good', 'Moderate', 'Unhealthy for Sensitive Groups',
    'Unhealthy', 'Very Unhealthy', 'Hazardous'
]

# Mapping Warna
quality_colors = {
    'Good': '#00e400',
    'Moderate': '#ffff00',
    'Unhealthy for Sensitive Groups': '#ff7e00',
    'Unhealthy': '#ff0000',
    'Very Unhealthy': '#8f3f97',
    'Hazardous': '#7e0023'
}

# --- 3. Pembuatan Tabel & Sorting ---
# Buat crosstab (persentase)
district_quality = pd.crosstab(
    air_quality_df['station'],
    air_quality_df['air_quality'],
    normalize='index'
) * 100

# Reindex kolom: Memastikan urutan warnanya Benar (Good -> Hazardous)
# Jika ada kategori yang 0%, kita isi dengan 0 agar tidak error
district_quality = district_quality.reindex(columns=category_order, fill_value=0)

# SORTING MAGIC: Urutkan Distrik dari yang 'Hazardous'-nya paling tinggi (Paling Kiri)
district_quality = district_quality.sort_values(by='Hazardous', ascending=False)

# --- 4. Plotting Visualisasi ---
plt.figure(figsize=(15, 9)) # Ukuran diperbesar agar lega

# Plot Stacked Bar
ax = district_quality.plot(
    kind='bar',
    stacked=True,
    color=[quality_colors[col] for col in district_quality.columns],
    width=0.85,          # Batang lebih tebal biar warnanya jelas
    edgecolor='black',   # Garis pinggir hitam tipis agar batas antar warna tegas
    linewidth=0.5,
    figsize=(15, 9)
)

# Judul dan Label
plt.title('Distribusi Kualitas Udara per Distrik', fontsize=16, pad=20, fontweight='bold')
plt.xlabel('Distrik', fontsize=12)
plt.ylabel('Persentase(%)', fontsize=12)

# Legend ditaruh di bawah agar tidak menutupi grafik
plt.legend(
    title='Kategori Kualitas Udara',
    bbox_to_anchor=(0.5, -0.25), # Posisi di bawah chart
    loc='upper center',
    ncol=6, # Berjejer ke samping
    fontsize=10
)

# Grid dan Layout
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()

# Tampilkan
plt.show()
```

## Analisis Lanjutan

Analisis clustering menggunakan beberapa parameter yang relevan dari data kualitas udara tersebut. Untuk kasus ini, kita bisa menggunakan parameter-parameter seperti PM2.5, PM10, SO2, NO2, CO, dan O3 yang merupakan indikator utama kualitas udara.

```python
def get_air_quality_stats(air_quality_df):
    # Pilih parameter yang akan digunakan
    parameters = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

    # Handling missing values
    for param in parameters:
        air_quality_df[param].fillna(air_quality_df[param].mean(), inplace=True)

    # Manual Grouping untuk PM2.5
    def categorize_pm25(value):
        if value <= 12:
            return 'Good'
        elif value <= 35.4:
            return 'Moderate'
        elif value <= 55.4:
            return 'Unhealthy for sensitive groups'
        elif value <= 150.4:
            return 'Unhealthy'
        elif value <= 250.4:
            return 'Very unhealthy'
        else:
            return 'Hazardous'

    air_quality_df['PM2.5_Category'] = air_quality_df['PM2.5'].apply(categorize_pm25)

    # Binning untuk parameter lain
    for param in parameters[1:]:  # Skip PM2.5 karena sudah dikategorikan
        air_quality_df[f'{param}_Binned'] = pd.qcut(air_quality_df[param], q=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'], duplicates='drop')

    # Statistik per kategori PM2.5
    stats_by_category = air_quality_df.groupby('PM2.5_Category')[parameters].agg(['mean', 'std', 'count'])

    # Statistik per distrik
    district_summary = air_quality_df.groupby('station')[parameters].agg(['mean', 'max', 'min'])

    return stats_by_category, district_summary

# Jalankan analisis
category_stats, district_stats = get_air_quality_stats(air_quality_df)

# Tampilkan hasil
print("\nStatistik per Kategori PM2.5:")
print(category_stats)
print("\nStatistik per Distrik:")
print(district_stats)
```
